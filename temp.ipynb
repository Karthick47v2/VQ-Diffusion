```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import textwrap
import glob
```

```python
paths = glob.glob('RESULT_PRETRAINED/*')

n_rows = 3
n_cols = 3

num_images = len(paths)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 10))

for i, path in enumerate(paths):
    img = Image.open(os.path.join(path, '000000.png'))
    row = i // n_cols
    col = i % n_cols
    axes[row, col].imshow(img)
    wrapped_title = '\n'.join(textwrap.wrap(path.split('/')[-1], width=40))
    axes[row, col].set_title(wrapped_title, fontsize=8)
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
